```python
import pandas as pd
df = pd.DataFrame({
    'Rank' : [1,2,3,4,5],
    'Employee': ['Ali', 'Bobur', 'Sardor', 'Jasur', 'Diyor'],
    'Salary': [5000, 4500, 5500, 6000, 4700]
})

# Shift bilan oldingi va keyingi qiymatlarni olish
df['Previous_Salary'] = df['Salary'].shift(1)
df['Next_Salary'] = df['Salary'].shift(-1)
df['Difference'] = df['Next_Salary'] - df['Previous_Salary'] 
print(df)
```

```python
df = pd.DataFrame({
    'Employee': ['Ali', 'Bobur', 'Sardor', 'Jasur', 'Diyor', 'Shoxrux'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'Finance'],
    'Salary': [5000, 4500, 5500, 6000, 4700, 6200],
    'JoinDate': pd.to_datetime(['2024-01-05', '2024-02-15', '2024-01-20', '2024-02-10', '2024-03-01', '2024-01-25'])
})

df['Month'] = df['JoinDate'].dt.strftime('%Y-%m')  # Extract year-month

# Pivot table: Bo‘lim + oy bo‘yicha o‘rtacha maosh
pivot = df.pivot_table(index='Department', columns='Month', values='Salary', aggfunc='mean', fill_value=0)

print(pivot)
```
