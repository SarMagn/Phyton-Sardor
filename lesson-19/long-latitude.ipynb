```python
import pyodbc
import requests
import time

# SQL Server'ga bogâ€˜lanish
conn = pyodbc.connect(
    "DRIVER={SQL Server};"
    "SERVER=Sardors-laptop\\SQLSERVER2025;"  # Ikkita backslash kerak
    "DATABASE=Banking_System;"
)
cursor = conn.cursor()

# Latitude va Longitude olish funksiyasi
def get_lat_lon_osm(address):
    url = f"https://nominatim.openstreetmap.org/search?q={address}&format=json"
    
    try:
        response = requests.get(url, timeout=10)  # 10 sekund timeout
        response.raise_for_status()  # Agar soâ€˜rovda xatolik boâ€˜lsa, chiqaradi
        
        data = response.json()
        if data:  # Maâ€™lumot boâ€˜lsa
            return data[0]["lat"], data[0]["lon"]
        else:
            return None, None  # Agar API hech narsa qaytarmasa
        
    except requests.exceptions.RequestException as e:
        print(f"ğŸŒ API ERROR: {e}")
        return None, None  # API xatosi boâ€˜lsa, `None` qaytaramiz

# SQL'dan barcha mijozlarning manzillarini olish
cursor.execute("""
    SELECT CustomerID, Address 
    FROM core.Customers 
    WHERE Latitude IS NULL OR Longitude IS NULL
""")
customers = cursor.fetchall()

# Har bir mijoz uchun manzillarni yangilash
for customer in customers:
    customer_id, address = customer
    lat, lon = get_lat_lon_osm(address)
    
    if lat and lon:
        cursor.execute("""
            UPDATE core.Customers 
            SET Latitude = ?, Longitude = ? 
            WHERE CustomerID = ?
        """, (lat, lon, customer_id))
        conn.commit()
        print(f"âœ… {address} => Latitude: {lat}, Longitude: {lon} yangilandi.")
    else:
        print(f"âŒ {address} uchun koordinata topilmadi yoki notoâ€˜gâ€˜ri manzil.")

    time.sleep(1)  # API bloklanmasligi uchun 1 soniya kutish

# Bogâ€˜lanishni yopish
cursor.close()
conn.close()

print("ğŸ‰ Barcha manzillar yangilandi!")
```

```python
import osmium

class HighwayHandler(osmium.SimpleHandler):
    def __init__(self):
        super(HighwayHandler, self).__init__()
        self.highways = []

    def way(self, w):
        if 'highway' in w.tags:
            self.highways.append([w.id, w.tags['highway'], w.nodes])

handler = HighwayHandler()
handler.apply_file(r"c:\Downloads\uzbekistan-latest.osm.pbf")

import csv
with open('highways.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['id', 'highway_type', 'nodes'])
    for h in handler.highways:
        writer.writerow(h)
```

```python
import osmium

class AddressHandler(osmium.SimpleHandler):
    def __init__(self):
        super(AddressHandler, self).__init__()
        self.addresses = []

    def node(self, n):
        # Tugunlardan (nodes) manzil maÊ¼lumotlarini olish
        if 'addr:street' in n.tags and 'addr:housenumber' in n.tags:
            address = {
                'id': n.id,
                'street': n.tags.get('addr:street'),
                'housenumber': n.tags.get('addr:housenumber'),
                'city': n.tags.get('addr:city', ''),
                'postcode': n.tags.get('addr:postcode', ''),
                'latitude': n.location.lat,
                'longitude': n.location.lon
            }
            self.addresses.append(address)

    def way(self, w):
        # YoÊ»llardan (ways) manzil maÊ¼lumotlarini olish
        if 'addr:street' in w.tags and 'addr:housenumber' in w.tags:
            address = {
                'id': w.id,
                'street': w.tags.get('addr:street'),
                'housenumber': w.tags.get('addr:housenumber'),
                'city': w.tags.get('addr:city', ''),
                'postcode': w.tags.get('addr:postcode', ''),
                'latitude': '',  # YoÊ»llar uchun kenglik/uzunlik maÊ¼lumoti yoÊ»q
                'longitude': ''
            }
            self.addresses.append(address)

# OSM faylini yuklash
handler = AddressHandler()
handler.apply_file(r"c:\Downloads\uzbekistan-latest.osm.pbf")

# Natijalarni CSV fayliga saqlash
import csv
with open('addresses.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['id', 'street', 'housenumber', 'city', 'postcode', 'latitude', 'longitude']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for address in handler.addresses:
        writer.writerow(address)

print(f"{len(handler.addresses)} ta manzil maÊ¼lumoti topildi va 'addresses.csv' fayliga saqlandi.")
```
