```python

data ='''{"id": 1, 
"name": "John", 
"age": 25,
"pets": ["dog", "cat", "birs"],
"phonenumber" : 
{
    "Phone" : "466-46-46",
    "HomePhone" : "456-98-32"
},

"familymembers": 
    [
        {
            "id": 100,
            "name" : "Josh",
            "relative" : "Father"
        },
        {
            "id": 101,
            "name" : "Sara",
            "relative" : "Mother"
        },
        {
            "id": 102,
            "name" : "Tom",
            "relative" : "Brother"
        },
        {
            "id": 103,
            "name" : "Anna",
            "relative" : "Sister"
        }





    ]

}'''
```

```python
import json

datar = json.loads(data)
datar

columns = ','.join(datar["familymembers"][0].keys()) + '\n'
# print(columns)

with open("students.csv", "w") as d: 
    d.write(columns) 
    for i in datar['familymembers']:
        values = ','.join(map(str,i.values())) + '\n'
        d.write(values)






   
```
