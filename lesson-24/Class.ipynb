```python
(flights
 .groupby(['ORG_AIR','DEST_AIR'])
 .agg({'CANCELLED':'sum',
       'AIR_TIME':'mean'}))
 
```

```python
(flights
 .groupby(['ORG_AIR','DEST_AIR'])
 .agg({'CANCELLED':['sum','mean','count'],
       'AIR_TIME':'mean'}))
```

```python
(flights
 .groupby(['ORG_AIR','DEST_AIR'])
 .agg(sum_cancelled = pd.NamedAgg(column='CANCELLED',aggfunc='sum'),
      mean_cancelled = pd.NamedAgg(column='CANCELLED',aggfunc='mean'),
      count_cancelled = pd.NamedAgg(column='CANCELLED',aggfunc='count'),
      mean_air_time = pd.NamedAgg(column='AIR_TIME',aggfunc='mean'),
      min_air_time = pd.NamedAgg(column='AIR_TIME',aggfunc='min'),
      max_air_time = pd.NamedAgg(column='AIR_TIME',aggfunc='max')))
```

```python
airline_info = (
    flights
    .groupby(['AIRLINE','WEEKDAY'])
    .agg({'DIST':['sum','mean'],
          'ARR_DELAY':['min','max']})
          .astype(int)
)
airline_info
```
